```python
sns.countplot(x='Species', data=df, )
```

```python
#1.Comparing variables by scatterplot

#a) Comparing SepalLengthC and SepalWidthCm
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm',
                hue='Species', data=data)
```

```python
#b) Comparing Petal Length and Petal Width
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm',
                hue='Species', data=data)
```

```python
#2.Pairplot
sns.pairplot(data)
```

```python
#3.Plotting Histogram
plt.hist(data['PetalLengthCm'])
plt.hist(data['SepalLengthCm'])
plt.hist(data['PetalWidthCm'])
plt.hist(data['SepalWidthCm'])
```

```python
#4.Displot
sns.distplot(a=data['PetalWidthCm'], bins=40, color='m')
```

```python
#6. Heatmap
sns.heatmap(data.corr(method='pearson').drop(
['Id'], axis=1).drop(['Id'], axis=0),
			annot = True);

plt.show()
```

```python
#7.Boxplot
sns.boxplot(x='Species',y='SepalWidthCm',data=data ,palette='YlGnBu')
```

```python
#8.Handling Outliers

sns.boxplot(x='SepalWidthCm', data=data)
```

```python
#9.3D plot
import plotly.express as px
fig = px.scatter_3d(data, x='SepalLengthCm', y='SepalWidthCm', z='PetalWidthCm')
fig.show()
```

```python
#10.Violonplot
sns.violinplot(x='Species', y='PetalLengthCm', data=data, palette='OrRd')
```

```python
#11.Lineplot
plt.plot(data["PetalLengthCm"])
plt.plot(data["PetalWidthCm"])
plt.xlabel('PetalLengthCm')
plt.ylabel('PetaWidthCm')
plt.show()
```

```python
#12.Clustermap
data_x=data.drop("Species",axis=1)
sns.clustermap(data_x, cmap="magma")
```

```python
#13.Regression Plot
sns.regplot(x='PetalWidthCm', y='PetalLengthCm', data=data)
```
